```python
import pandas as pd
import numpy as np
df = pd.read_csv('D:\PycharmProjects\Jupyter-Notebooks-master\Datasets\BL-Flickr-Images-Book2.csv')
df.head()
```

```python
to_drop = ['Edition Statement',
          'Corporate Author',
           'Corporate Contributors',
           'Former owner',
           'Engraver',
           'Contributors',
           'Issuance type',
           'Shelfmarks']
```

```python
df.drop(columns=to_drop,inplace=True)
```

```python
df.head()
```

```python
df['Identifier'].is_unique
```

```python
df.set_index('Identifier', inplace=True)
```

```python
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
```

```python
df['Date of Publication'].head()
```

```python
df.head()
```

```python
pub = df['Place of Publication']
london = pub.str.contains('London')
oxford = pub.str.contains('Oxford')
df['Place of Publication'] = np.where(london,'London', np.where(oxford,'Oxford',pub.str.replace('-',' ')))
df.head()
```
